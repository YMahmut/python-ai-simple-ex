```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import chain
import pandas as pd
data=pd.read_excel('file path.xlsx','excell file page number')
data.head()
veri_ayrilmamis=data.to_numpy()
veri_ayrilmamis=np.array(veri_ayrilmamis)


def func_sutun_kategorilestirme(future_alan_param):
    olan_listesi=[]
    yeni_a=[]
    tur_no=1
    dict_tur_no_tut={}
    for a_kategori in range(len(future_alan_param)):
        var_mi=future_alan_param[a_kategori:(a_kategori + 1)]
        if var_mi not in olan_listesi:
            olan_listesi.append(var_mi)
            yeni_a.append(tur_no)
            dict_tur_no_tut[tur_no]=var_mi
            tur_no +=1
        else:
            yeni_a.append(list(dict_tur_no_tut.keys())[list(dict_tur_no_tut.values()).index(var_mi)] )
    return yeni_a
kategorilestirme=veri_ayrilmamis[:,(len(veri_ayrilmamis[0])-1):len(veri_ayrilmamis[0])]
train_y=func_sutun_kategorilestirme(kategorilestirme)
veri_ayrilmamis[:,(len(veri_ayrilmamis[0])-1):len(veri_ayrilmamis[0])]=np.array(train_y).reshape((-1,1))


test_veri=veri_ayrilmamis[0:1]
train_veri=veri_ayrilmamis[1:2]
for bol1 in range(len(veri_ayrilmamis)-2):
    siradaki=veri_ayrilmamis[(bol1+2):(bol1+3)]
    if ((bol1+2)%3)==0:
        test_veri=np.concatenate((test_veri, siradaki))
    else:
        train_veri=np.concatenate((train_veri, siradaki))


        
a=np.array(train_veri)
b=np.array(test_veri)
#b_kararli=np.array(test_veri)
#b=b_kararli[:,0:(len(b_kararli[0])-1)]
a_ARR=np.array(a)
b_ARR=np.array(b)

optimum_K_elemn=8
print("              En Yakin Komsuluk Sayisi 'K' = ",(optimum_K_elemn))
Uzaklık_=[]
for eleman2 in range(len(b)):
    mesafeler_gcc_listesi=[]
    for eleman_satir in range(len(a)):
        toplam=0
        L_uzklk_sinf=[]
        for eleman_sutun in range(len(a[0])-1):
            farkin_karesi = (a[eleman_satir][eleman_sutun]-b[eleman2][eleman_sutun])**2
            toplam += farkin_karesi
        kok=toplam**(1/2)
        L_uzklk_sinf.append(kok)
        L_uzklk_sinf.append(a[eleman_satir][-1])
        mesafeler_gcc_listesi.append(L_uzklk_sinf)
    mesafeler_gcc_listesi_ARR=np.array(mesafeler_gcc_listesi)
    sorted_array = mesafeler_gcc_listesi_ARR[np.argsort(mesafeler_gcc_listesi_ARR[:, 0])] 
    Uzaklık_.append(sorted_array)
Uzaklık_ARR=np.array(Uzaklık_)

#print("uzaklik:\n",Uzaklık_ARR)
Tahmin_Listesi=[]
for komsuluk_eleman1 in range(len(Uzaklık_)):
    komsuluk_Listesi = ((Uzaklık_[komsuluk_eleman1][0:(optimum_K_elemn +1)])[:,1:2]).tolist()
    dict_Frkns ={}
    komsuluk_Listesi=list(chain(*komsuluk_Listesi))
    for i in komsuluk_Listesi:
        dict_Frkns[i] = komsuluk_Listesi.count(i)
    Tahmin=int(max(dict_Frkns, key=dict_Frkns.get))
    Tahmin_Listesi.append(Tahmin)
    #print(b[komsuluk_eleman1],"   ",Tahmin,".   Sınıf")
print("---------------------------------------------------------------------------")

Tahmin_L_ARR_TT=((np.array(Tahmin_Listesi)).reshape((-1,1)))
Gercek_L_ARR=(b_ARR[:,(len(b_ARR[0])-1):(len(b_ARR[0]))])
sonuclistesi=(Gercek_L_ARR==Tahmin_L_ARR_TT)
print("Doğru Sınıflama Sayisi",(sonuclistesi.tolist().count([True])))
print("Yanlış Sınıflama Sayisi",(sonuclistesi.tolist().count([False])))


def bir_oznitelikteki_tur_sayisi(feature_):
    a=feature_.tolist()
    liste1=[]
    for b in range(len(a)):
        count = 0
        for element in a:
            count += element.count((b+1))
        if count!=0:
            liste1.append((b+1))
    return liste1
tur_listesi_Dec_Test=bir_oznitelikteki_tur_sayisi(Gercek_L_ARR)

tur_listesi_Dec_Test

CofusionListesi=[]
for confkt in range(len(tur_listesi_Dec_Test)):
    strconfK=(confkt+1)
    strconfK=str(strconfK)
    strconfK="gercek"+strconfK
    strconfK=[]
    CofusionListesi.append(strconfK)
Son_Conf_Listesi=[]
for ss1 in range(len(Tahmin_L_ARR_TT)):
    if (Gercek_L_ARR[ss1][0]==Tahmin_L_ARR_TT[ss1][0]):
        for cnmk in range(len(CofusionListesi)):
            if (cnmk+1)==Gercek_L_ARR[ss1][0]:
                CofusionListesi[cnmk].append(int(str(Gercek_L_ARR[ss1][0])+str(Tahmin_L_ARR_TT[ss1][0])))
            else:
                continue
    else:
        for cnmk in range(len(CofusionListesi)):
            if (cnmk+1)==Gercek_L_ARR[ss1][0]:
                CofusionListesi[cnmk].append(int(str(Gercek_L_ARR[ss1][0])+str(Tahmin_L_ARR_TT[ss1][0])))
            else:
                continue

Son_Conf_Listesi_S=list(chain(*CofusionListesi))

teams_list01 = []
for lll1 in range(len(tur_listesi_Dec_Test)):
    for lll2 in range(len(tur_listesi_Dec_Test)):
        teams_list01.append([int(str(lll1+1)+str(lll2+1))])

liste1234=[]
liste12345=[]
dict12345={}
for abcd1 in range(len(Son_Conf_Listesi_S)):
    if Son_Conf_Listesi_S[abcd1] not in liste1234:
        liste1234.append(Son_Conf_Listesi_S[abcd1])
        gecici_scf=Son_Conf_Listesi_S.count(liste1234[-1])
        liste12345.append(gecici_scf)
        dict12345[liste1234[-1]]=liste12345[-1]
    if abcd1==((len(Son_Conf_Listesi_S))-1):
        if len(teams_list01) != len(liste1234):
            li1=liste1234
            li2=list(chain(*teams_list01))
            li3=list(list(set(li1)-set(li2)) + list(set(li2)-set(li1)))
            for li330 in range(len(li3)):
                li1.append(li3[li330])
                liste12345.append(0)
                dict12345[liste1234[-1]]=liste12345[-1]
            liste1234=li1
for div_by_zero_krtr_elmn in range(len(tur_listesi_Dec_Test)):
    if dict12345[int(str(div_by_zero_krtr_elmn+1) + str(div_by_zero_krtr_elmn+1))]==0:
        dict12345[int(str(div_by_zero_krtr_elmn+1) + str(div_by_zero_krtr_elmn+1))] =1
        print("dvdbyzr")  
dict12345        
teams_list = []
tablo_list=[]
for lll1 in range(len(tur_listesi_Dec_Test)):
    teams_list.append(str(tur_listesi_Dec_Test[lll1]))
    tablo_list_gec=[]
    for lll2 in range(len(tur_listesi_Dec_Test)):
        tablo_list_gec.append(dict12345[int(str(lll1+1)+str(lll2+1))])
    tablo_list.append(tablo_list_gec)


#print(teams_list)
#print(tablo_list)
cokSinifli_Confusion_Matrix=pd.DataFrame(tablo_list, teams_list, teams_list)
print("\n------------------------------\n")
print("confusion matrix ")
display(cokSinifli_Confusion_Matrix)

TrPz_Liste=[]
FlPz_Liste=[]
FlNg_Liste=[]

for cm_elm_bul1 in range(len(teams_list)):
    FlPz_Liste_gecici=[]
    FlNg_Liste_gecici=[]
    
    for cm_elm_bul2 in range(len(tablo_list)):
        
        if ((1+cm_elm_bul1)==(1+cm_elm_bul2)):
            TrPz=dict12345.get(int(str(1+cm_elm_bul1)+str(1+cm_elm_bul2)),0)
            TrPz_Liste.append(TrPz)
        else:
            
            FlPz_gcc=dict12345.get(int(str(1+cm_elm_bul1)+str(1+cm_elm_bul2)),0)
            FlPz_Liste_gecici.append(FlPz_gcc)
            FlNg_gcc=dict12345.get(int(str(1+cm_elm_bul2)+str(1+cm_elm_bul1)),0)
            FlNg_Liste_gecici.append(FlNg_gcc)
            
    FlPz=sum(FlPz_Liste_gecici)
    FlPz_Liste.append(FlPz)
    FlNg=sum(FlNg_Liste_gecici)
    FlNg_Liste.append(FlNg)



#true pozitif , false pozitif ...
print("--------------------------")
print("TrPz_Liste: ",TrPz_Liste)
print("FlPz_Liste: ",FlPz_Liste)
print("FlNg_Liste: ",FlNg_Liste)

TrNg_Liste=[]
for cm_TrNg_elm_bul1 in range(len(teams_list)):
    tablo_list_cpy=tablo_list[:]
    TrNg = np.delete(tablo_list_cpy, (cm_TrNg_elm_bul1), axis=0)
    TrNg = np.delete(TrNg, (cm_TrNg_elm_bul1), axis=1)
    TrNg_Liste.append((TrNg.sum()))
print("TrNg_Liste: ",TrNg_Liste)
print("______________________________\n")

np.set_printoptions(precision=4)  # ondalık basamak hassasiyeti 4 basamak olsun
mac_f1_liste=[]
for metricHesapla1 in range(len(teams_list)):
    
    precision_class=TrPz_Liste[metricHesapla1]/(TrPz_Liste[metricHesapla1]+FlPz_Liste[metricHesapla1])
    print("precision_class"+str((metricHesapla1+1)),":  ",round(precision_class,2))
    
    recall_class=TrPz_Liste[metricHesapla1]/(TrPz_Liste[metricHesapla1]+FlNg_Liste[metricHesapla1])
    print("recall_class"+str((metricHesapla1+1))," :    ",round(recall_class,2))
    
    F_1_Score_class=2*precision_class*recall_class/(precision_class+recall_class)
    print("F_1_Score_class"+str((metricHesapla1+1)),":  ",round(F_1_Score_class,2))
    
    mac_f1_liste.append(F_1_Score_class)
    
    specification_class=TrNg_Liste[metricHesapla1]/(TrNg_Liste[metricHesapla1]+FlPz_Liste[metricHesapla1])
    print("specification_class"+str((metricHesapla1+1)),":  ",round(specification_class,2))
    toplam=0
    for acc_hesapla2 in range(len(tablo_list[0])):
        tp=tablo_list[metricHesapla1][metricHesapla1]
        toplam+=tablo_list[acc_hesapla2][metricHesapla1]
    Accuracy_class=tp/toplam
    print("Accuracy_class",str(metricHesapla1+1),":  ",round(Accuracy_class,2))
    print("---------------")
#microMetrikler
Mic_precision_=sum(TrPz_Liste)/(sum(TrPz_Liste)+sum(FlPz_Liste))
Mic_recall_=sum(TrPz_Liste)/(sum(TrPz_Liste)+sum(FlNg_Liste))
Mic_F_1_Score_=2*Mic_precision_*Mic_recall_/(Mic_precision_+Mic_recall_)
Mic_specification_=sum(TrNg_Liste)/(sum(TrNg_Liste)+sum(FlPz_Liste))
Mic_Accuracy_=sum(TrPz_Liste)/(sum((np.array(tablo_list)).sum(axis=0)))

print("Mic_F_1_Score_  : ",round(Mic_F_1_Score_,2))
print("")
Mac_F_1_Score_=round((sum(mac_f1_liste)/(len(teams_list))),2)
print("Mac_F_1_Score_  : ",Mac_F_1_Score_)
print("")
agirlikli_F_1_Score_liste=[]
for agirliklihesap1 in range(len(teams_list)):
    agirlikli_F_1_Score_liste.append((mac_f1_liste[agirliklihesap1]*(((np.array(tablo_list)).sum(axis=0)).tolist()[agirliklihesap1])))
agirlikli_F_1_Score_=round(sum(agirlikli_F_1_Score_liste)/(sum(((np.array(tablo_list)).sum(axis=0)).tolist())),2)
print("agirlikli_F_1_Score_  : ",agirlikli_F_1_Score_)
```
