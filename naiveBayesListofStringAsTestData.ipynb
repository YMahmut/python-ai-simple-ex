```python

import numpy as np
import pandas as pd
import math

veri_ayrilmamis= np.array(
               [["sunny","hot","high","weak","no"]\
               ,["sunny","hot","high","strong","no"]\
               ,["overcast","hot","high","weak","yes"]\
               ,["rainy","mild","high","weak","yes"]\
               ,["rainy","cool","normal","weak","yes"]\
               ,["rainy","cool","normal","strong","no"]\
               ,["overcast","cool","normal","strong","yes"]\
               ,["sunny","mild","high","weak","no"]\
               ,["sunny","cool","normal","weak","yes"]\
               ,["rainy","mild","normal","weak","yes"]\
               ,["sunny","mild","normal","strong","yes"]\
               ,["overcast","mild","high","strong","yes"]\
               ,["overcast","hot","normal","weak","yes"]\
               ,["rainy","mild","high","strong","no"]\
              ])
```

```python
FEATURES={}
def func_sutun_kategorilestirme(future_alan_param):
    olan_listesi=[]
    yeni_a=[]
    tur_no=1
    dict_tur_no_tut={}
    for a_kategori in range(len(future_alan_param)):
        var_mi=future_alan_param[a_kategori:(a_kategori + 1)]
        if var_mi not in olan_listesi:
            FEATURES[str(var_mi[0][0])]=tur_no
            olan_listesi.append(var_mi)
            yeni_a.append(tur_no)
            dict_tur_no_tut[tur_no]=var_mi
            tur_no +=1
        else:
            yeni_a.append(list(dict_tur_no_tut.keys())[list(dict_tur_no_tut.values()).index(var_mi)] )
    return yeni_a
kategorilestirme=veri_ayrilmamis[:,(len(veri_ayrilmamis[0])-1):len(veri_ayrilmamis[0])]
#print("kategorilestirme",kategorilestirme)
train_y=func_sutun_kategorilestirme(kategorilestirme)
#print("train_y",train_y)
```

```python
kategorikOznitelikler=[];
for featureColumn in range(len(veri_ayrilmamis[0])-1):
    kategorilestirmex1=veri_ayrilmamis[:,featureColumn:featureColumn+1]
    train_x=func_sutun_kategorilestirme(kategorilestirmex1)
    kategorikOznitelik=np.array(train_x);
    kategorikOznitelikler.append(kategorikOznitelik)
kategorikOznitelikler.append(np.array(train_y))
veri=np.array(kategorikOznitelikler).transpose()
print("veri \n", veri)
print("")
print("öznitelikler\n",FEATURES)
```
